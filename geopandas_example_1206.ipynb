```python
import pandas as pd
from shapely.geometry import Point
```

```python
volcano_file = "C:/Users/SKW/Desktop/dataset/volcano of world.xlsx"
volcano = pd.read_excel(volcano_file)
volcano.head()
```

```python
volcano_df = volcano[['Latitude', 'Longitude']]
volcano_df
```

```python
volcano_df['coordinates'] = volcano_df.apply(lambda row: Point(row['Longitude'], row['Latitude']), axis=1)
gdf = gpd.GeoDataFrame(volcano_df, geometry='coordinates')
gdf
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
base = world.plot(figsize=(15, 15))
gdf.plot(ax = base, color = 'red', markersize = 10)
base.set_title("Volcano in the World", fontsize = 20)
base.set_axis_off()
plt.show()
```
